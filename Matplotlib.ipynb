```python
"""
    Scatter plot is type of plot in which data is represented thorugh data points. This points are plotted across x-axis and y-axis. """
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Title of the plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
"""
    A bar plot is used to plot data in form of bar chart where the charts describes the comparisons between the discrete categories.
    It is used to compare sets of data between different groups at once. """
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```
