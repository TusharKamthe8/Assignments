```python
from flask import Flask
app = Flask(__name__)
@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
"""
    1)url_for() function is used in flask for url building. """

from flask import Flask, url_for
app = Flask(__name__)

@app.route('/user/<name>')
def hello(name):
    return 'Hello %s'%name

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
"""
    url_for is used for url building. In the above code we have a function hello which takes name as argument and then gives result.
    In app.route we are naming the function as '/user' and then passing arguement'<name>' which will take the argument and give it to the function.
    Now when we use the url to open this application we put the name of function and the input we want side by side as per the protocol.
    In below image we can see the output with the changed url as per the input. """
```
